```python
import pandas as pd

df = pd.read_csv("retail_sales_dataset.csv")  # apna exact naam daalo
df.head()
```

```python
total_sales = df["Sales"].sum()
print("Total Sales:", total_sales)
```

```python
print(df.columns)
```

```python
total_sales = df["Total Amount"].sum()
print("Total Sales:", total_sales)
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df["Year"] = df["Date"].dt.year

yearly_sales = df.groupby("Year")["Total Amount"].sum()

yearly_sales.plot(kind="line", marker="o")
```

```python
top_category = df.groupby("Product Category")["Total Amount"].sum().sort_values(ascending=False)

top_category.plot(kind="bar")
```

```python
gender_sales = df.groupby("Gender")["Total Amount"].sum()

gender_sales.plot(kind="bar")
```
